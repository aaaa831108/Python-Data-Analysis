# 使用函式append() 合併與新增資料
## 將兩個DataFrame合併在一起

```python
import pandas as pd

df1 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
                    'B':['B0', 'B1', 'B2', 'B3'],
                    'C':['C0', 'C1', 'C2', 'C3'],
                    'D':['D0', 'D1', 'D2', 'D3']},
                   index = [0, 1, 2, 3])
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6', 'A7'],
                    'B':['B4', 'B5', 'B6', 'B7'],
                    'C':['C4', 'C5', 'C6', 'C7'],
                    'D':['D4', 'D5', 'D6', 'D7']},
                   index = [4, 5, 6, 7])

df1.append(df2)
```

## 合併兩個以上的DataFrame

```python
df3 = pd.DataFrame({'A':['A8', 'A9', 'A10', 'A11'],
                    'B':['B8', 'B9', 'B10', 'B11'],
                    'C':['C8', 'C9', 'C10', 'C11'],
                    'D':['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
df1.append([df2, df3])
```

## 使用參數ignore_index，建立新的index

```python
df1
```

```python
df4 = pd.DataFrame({'B':['B2', 'B3', 'B6', 'B7'],
                    'D':['D2', 'D3', 'D6', 'D7'],
                    'F':['F2', 'F3', 'F6', 'F7']},
                   index = [2, 3, 6, 7])
df1.append(df4)
```

```python
df1.append(df4, ignore_index=True)
```

## 新增一筆資料的兩種方式
* loc\[索引值\] = \[資料1, 資料2...\]
* append({欄位1:資料1, 欄位2:資料2, ....})

```python
df1
```

```python
df1.loc[4] = ['A4', 'B4', 'C4', 'D4']
df1
```

```python
df1 = df1.append({'A':'A5', 'B':'B5', 'C':'C5', 'D':'D5'}, ignore_index=True)
df1
```

```python
df1 = df1.append({'A':'A5', 'B':'B5', 'C':'C5', 'D':'D5'})
df1
```
